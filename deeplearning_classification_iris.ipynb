```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
dictionary = {'setosa':0,'versicolor':1,'virginica':2}
df['species']=df['species'].map(dictionary)
df.head()
```

```python
x = df.drop('species',axis=1)
```

```python
import tensorflow as tf
import numpy as np
```

```python
y = tf.keras.utils.to_categorical(df['species'])   # catagorical is the function whicich converts bigger number to [1,0,0,0] kind of number
```

```python
inputlayer = tf.keras.layers.Dense(units = 4,input_dim = 4)
layer1 = tf.keras.layers.Dense(units =32)
layer2 = tf.keras.layers.Dense(units =32)
layer3 = tf.keras.layers.Dense(units =64)
layer4 = tf.keras.layers.Dense(units =64)
layer5 = tf.keras.layers.Dense(units =128)
layer6 = tf.keras.layers.Dense(units =64)
outputlayer = tf.keras.layers.Dense(units = 3, activation='softmax')
```

```python
model = tf.keras.Sequential()
model.add(inputlayer)
model.add(layer1)
model.add(layer2)
model.add(layer3)
model.add(layer4)
model.add(layer5)
model.add(layer6)
model.add(outputlayer)
```

```python
model.compile(optimizer = 'adam' , loss ='categorical_crossentropy' , metrics = ['accuracy'])
```

```python
model.fit(x,y,epochs=50)
```
