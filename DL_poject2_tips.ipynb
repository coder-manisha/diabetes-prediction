```python
import seaborn as sns
```

###### to get dataset name ('tips') : search in google :load dataset in seaborn documentation  ---> click on first link-->again click on the name dataset link --> search for related dataset name --> then write that name here .

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
dictionary1 = {'Female':0,'Male':1}
df['sex'] = df['sex'].map(dictionary1)
df.head()
```

```python
df['smoker'].unique()
```

```python
dictionary2 = {'No':0,'Yes':1}
df['smoker'] = df['smoker'].map(dictionary2)
df.head()
```

```python
df['day'].unique()
```

```python
dictionary3 = {'Sun':1,'Sat':2,'Thur':3,'Fri':4}
df['day']=df['day'].map(dictionary3)
df.head()
```

```python
df['time'].unique()
```

```python
dictionary4 = {'Dinner':0,'Lunch':1}
df['time'] = df['time'].map(dictionary4)
df.head()
```

###### x axis contains all columns except tip and y is assigned to tip because we have to find tip

```python
x = df.drop('tip',axis=1)
y = df['tip']
```

```python
import tensorflow as tf
import numpy as np
```

```python
#step1 : create neuron layers with neurons

layer1 = tf.keras.layers.Dense(units=4 , input_dim =6)
layer2 = tf.keras.layers.Dense(units=32)
layer3 = tf.keras.layers.Dense(units=32)
layer4 = tf.keras.layers.Dense(units=64)
layer5 = tf.keras.layers.Dense(units=64)
layer10 = tf.keras.layers.Dense(units=64)
layer6 = tf.keras.layers.Dense(units=128)
layer7 = tf.keras.layers.Dense(units=128)
layer8 = tf.keras.layers.Dense(units=64)
layer9 = tf.keras.layers.Dense(units=1)
```

```python
#step 2: add all layers tomodel
model = tf.keras.Sequential()
model.add(layer1)
model.add(layer2)
model.add(layer3)
model.add(layer4)
model.add(layer5)
model.add(layer6)
model.add(layer7)
model.add(layer8)
model.add(layer9)
model.add(layer10)
```

```python
#step 3: giveoptimizer and loss function to model
model.compile(optimizer='adam',loss='mean_squared_error',metrics = ['mean_squared_error'])
```

```python
model.fit(x,y,epochs=500)
```

```python
df.head()
```

```python
model.predict(np.array([[21.01,1,0,1,0,3]]))
```

```python
#more the layers will get most accurate answer
model.predict(np.array([[16.99,1.01,0,1,0,2]]))
```
